```python
import urllib2
connection = urllib2.urlopen('https://en.wikipedia.org/wiki/Lightweight_Java_Game_Library')
page = connection.read()
connection.close()
print(page)
```

```python
import zlib
compressedPage = zlib.compress(page)
print(compressedPage)
```

```python
import sys
originalSize = sys.getsizeof(page)
compressedSize = sys.getsizeof(compressedPage)
print("Original : %i" % originalSize + " bytes")
print("Compressed : %i" % compressedSize + " bytes")
print("%f%% size reduction" % (100.0-100.0*compressedSize/originalSize))
```
