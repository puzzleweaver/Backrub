```python
def is_valid_wordID(wordIDBytes):
    for i in wordIDBytes:
        if i != 0:
            return True
    return False

def forward_index_review():
    with open('barrels/forward_index.bin', 'r+b') as fp:
        fp.seek(0, 2)
        size = fp.tell()
        fp.seek(0, 0)
        while(fp.tell() < size):
            docID = struct.unpack("I", fp.read(4))[0]
            print "DOC ID: ", docID
            while is_valid_wordID(struct.unpack("BBB", fp.read(3))):
                fp.seek(-3, 1)
                hit_data = struct.unpack("I", fp.read(4))[0]
                num_hits = hit_data & 0xFF
                wordID = hit_data >>  24 & 0xFFFFFF
                print "WordID: ", wordID
                print "num_hits: ", num_hits
                print ""
                fp.seek(num_hits*2, 1)
            print "======================="

forward_index_review()
```
