```python
import requests
import random
import re
link = "https://en.wikipedia.org/wiki/Lightweight_Java_Game_Library"

def getLinkList(str):
    return re.findall("href=\"(.+?)\"", str)

for i in xrange(100):
    f = requests.get(link)
    links = getLinkList(f.text)
    while True:
        link = links[random.randint(0, len(links)-1)]
        if link[0] == "/" and len(getLinkList(requests.get("https://en.wikipedia.org" + link).text)) != 0:
            break
    link = "https://en.wikipedia.org" + link
    print("%i: "%len(links) + link)
```

```python
import binascii
def checksum(url):
    return binascii.crc32(url) & 0xffffffff #this keeps the results consistent across all platforms
```

```python
def requestPage(url):
    import urllib2
    connection = urllib2.urlopen('https://en.wikipedia.org/wiki/Lightweight_Java_Game_Library')
    page = connection.read()
    connection.close()
    return page

def parseLinks(url, page):
    import re
    endOfBase = url[9:].index('/')
    baseURL = url[0:endOfBase]
    uncheckedLinks = re.findall('href="([^"]*)"', page) #Parses all the links FROM the current page
    linksFrom = []
    for i in uncheckedLinks:
        if i[0:2] == "//": #if the only thing missing is "https" or another protocol
            linksFrom.append("https:" + i)
        elif i[0:1] == "/": #if the link is on the same site
            linksFrom.append(baseURL + i) #concat the site base URL to the rest of the URL
    return linksFrom

checksumToIndex = []
def putPageInDatabase(url):
    urlChecksum = checksum(url)
    if checksumToIndex.count(urlChecksum) == 0:
        checksumToIndex.append(urlChecksum)
def getUrlIndex(url):
    return checksumToIndex.index(checksum(url))

pageURL = "https://en.wikipedia.org/wiki/Lightweight_Java_Game_Library"
page = requestPage(pageURL)
links = parseLinks(pageURL, page)
for i in links:
    putPageInDatabase(i)
    print ("%i" %getUrlIndex(i) + ": " + i)
```
