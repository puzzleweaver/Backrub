```python
import requests
import random
import re
link = "https://en.wikipedia.org/wiki/Lightweight_Java_Game_Library"

def getLinkList(str):
    return re.findall("href=\"(.+?)\"", str)

for i in xrange(10):
    f = requests.get(link)
    links = getLinkList(f.text)
    while True:
        link = links[random.randint(0, len(links)-1)]
        if link[0] == "/" and len(getLinkList(requests.get("https://en.wikipedia.org" + link).text)) != 0:
            break
    link = "https://en.wikipedia.org" + link
    print("%i: "%len(links) + link)
```

```python
import binascii
def checksum(url):
    return binascii.crc32(url) & 0xffffffff #this keeps the results consistent across all platforms
```

```python
def requestPage(url):
    import urllib2
    try: 
        connection = urllib2.urlopen(url)
    except urllib2.HTTPError, e:
        print('HTTPError = ' + str(e.code))
        return ""
    page = connection.read()
    connection.close()
    return page

def parseLinks(url, page):
    import re
    import urlparse
    parse = urlparse.urlparse(url)
    baseURL = parse[0] + "://" + parse[1]
    #print("Base URL: " + baseURL)
    uncheckedLinks = re.findall('href="([^"]*)"', page) #Parses all the links FROM the current page
    linksFrom = []
    for i in uncheckedLinks:
        if i[0:2] == "//": #if the only thing missing is "https" or another protocol
            linksFrom.append(parse[0] + ":" + i)
        elif i[0:1] == "/": #if the link is on the same site
            linksFrom.append(baseURL + i) #concat the site base URL to the rest of the URL
    return linksFrom
```

```python
import zlib
import sys
import time
import struct

def compressPage(page):
    startTime = time.clock()
    compressedPage = zlib.compress(page)
    endTime = time.clock()
    originalSize = sys.getsizeof(page)
    compressedSize = sys.getsizeof(compressedPage)
    print("\tOriginal : %i" % originalSize + " bytes")
    print("\tCompressed : %i" % compressedSize + " bytes")
    print("\t%f%% size reduction" % (100.0-100.0*compressedSize/originalSize))
    print("\tCompression time: %f" % (endTime-startTime) + " seconds")
    return compressedPage

def decompressPage(compressedPage):
    startTime = time.clock()
    decompressedPage = zlib.decompress(compressedPage)
    endTime = time.clock()
    print("Decompression time: %f" % (endTime-startTime) + " seconds")
    return decompressedPage

def savePage(docID, URL, page):#we could do so much better with file.seek(byte)
    pageFile = open('D:\\poople\\page_repository\\%i' % docID, 'w')
    data = "%i" %docID + ",%i" %len(page) + "," + URL + "," + page
    compressedData = compressPage(data)
    pageFile.write(compressedData)
    pageFile.close()
#NOAH FIX THE FILESYSTEM STUFF
#savePage(0, pageURL, page)
```

```python
import math
import bisect

checksums = []
def putPageInDatabase(url):
    urlChecksum = checksum(url)
    cID = int(len(checksums)/2)
    delta = int(math.ceil(cID/2))
    if(len(checksums) == 0):
        checksums.append(urlChecksum)
    else:
        checksums.insert(bisect.bisect(checksums, urlChecksum), urlChecksum)
def getUrlIndex(url):
    return bisect.bisect(checksums, checksum(url))-1

pageURL = "https://en.wikipedia.org/wiki/Lightweight_Java_Game_Library"
page = requestPage(pageURL)
links = parseLinks(pageURL, page)
for i in links:
    putPageInDatabase(i)
    
for i in links:
    # these shouldn't print in order
    print("%i : " % checksum(i) + "%i : " % checksums[getUrlIndex(i)] + i)
```
